```python
# Dividir los valores de 'listed_in' en listas
netflix_df['listed_in_split'] = netflix_df['listed_in'].apply(lambda x: x.split(', '))

# Contar todas las categorías únicas
from collections import Counter
categories = Counter([genre for sublist in netflix_df['listed_in_split'] for genre in sublist])

# Filtrar las categorías principales (por ejemplo, las 20 más frecuentes)
top_categories = [category for category, count in categories.most_common(20)]

# Crear una nueva columna con categorías simplificadas
def map_to_top_categories(genres):
    mapped = [genre for genre in genres if genre in top_categories]
    return mapped if mapped else ['Otros']  # Si no pertenece a las principales, asignar 'Otros'

netflix_df['simplified_genres'] = netflix_df['listed_in_split'].apply(map_to_top_categories)

# Combinar las categorías en una sola cadena (opcional)
netflix_df['simplified_genres'] = netflix_df['simplified_genres'].apply(lambda x: ', '.join(x))

print(netflix_df[['listed_in', 'simplified_genres']].head())
```

```python
import pandas as pd

# Cargar el dataset
file_path = "/content/drive/MyDrive/TFM/Datasets/netflix_integracionv3.csv"
netflix_df = pd.read_csv(file_path)

# Asegurarnos de que existe la columna 'simplified_genres'
if 'simplified_genres' not in netflix_df.columns:
    # Crear 'simplified_genres' a partir de 'listed_in' si no existe
    netflix_df['listed_in_split'] = netflix_df['listed_in'].apply(lambda x: x.split(', '))

    # Contar todas las categorías únicas
    from collections import Counter
    categories = Counter([genre for sublist in netflix_df['listed_in_split'] for genre in sublist])

    # Filtrar las categorías principales (ejemplo: 20 más comunes)
    top_categories = [category for category, count in categories.most_common(20)]

    # Crear una nueva columna con categorías simplificadas
    def map_to_top_categories(genres):
        mapped = [genre for genre in genres if genre in top_categories]
        return mapped if mapped else ['Otros']  # Si no pertenece a las principales, asignar 'Otros'

    netflix_df['simplified_genres'] = netflix_df['listed_in_split'].apply(map_to_top_categories)
    netflix_df['simplified_genres'] = netflix_df['simplified_genres'].apply(lambda x: ', '.join(x))

# Crear la columna 'primary_two_genres' si no existe
if 'primary_two_genres' not in netflix_df.columns:
    def extract_top_two(genres):
        split_genres = genres.split(', ')
        return ', '.join(split_genres[:2])  # Tomar las primeras dos etiquetas

    netflix_df['primary_two_genres'] = netflix_df['simplified_genres'].apply(extract_top_two)

# Refinar combinaciones desordenadas en 'primary_two_genres'
netflix_df['primary_two_genres_refined'] = netflix_df['primary_two_genres'].apply(
    lambda x: ', '.join(sorted(x.split(', ')))
)

# Revisar la distribución refinada
refined_distribution = netflix_df['primary_two_genres_refined'].value_counts()

# Mostrar los resultados
print("Distribución refinada de combinaciones:")
print(refined_distribution.head())
print(f"\nNúmero total de combinaciones únicas refinadas: {len(refined_distribution)}")
```

```python
# Eliminar columnas intermedias
columns_to_drop = ['listed_in', 'simplified_genres', 'primary_two_genres']
netflix_df = netflix_df.drop(columns=columns_to_drop)

# Guardar el dataset limpio con la columna refinada
output_file_path = "/content/drive/MyDrive/TFM/Datasets/netflix_cleaned.csv"
netflix_df.to_csv(output_file_path, index=False)

print(f"Dataset limpio guardado en: {output_file_path}")
print("Columnas finales en el dataset:")
print(netflix_df.columns)
```

```python
import pandas as pd

# Cargar el dataset
file_path = "/content/drive/MyDrive/TFM/Datasets/netflix_sample_for_labeling.csv"
dataset = pd.read_csv(file_path)

# Convertir la columna 'emotion' a texto para las etiquetas
dataset['emotion'] = dataset['emotion'].astype(str)

# Lista de emociones
emotion_options = [
    "Happy", "Sad", "Nostalgic", "Excited", "Scared",
    "Angry", "Romantic", "Hopeful", "Funny", "Adventurous",
    "Suspenseful", "Inspiring", "Heartwarming"
]

# Función para etiquetar manualmente con etiquetas compuestas
def etiquetar_manual(df):
    print("Inicio del proceso de etiquetado manual. Escribe '0' para guardar y salir.\n")
    for index, row in df.iterrows():
        if row['emotion'] == "nan" or row['emotion'] == "":  # Solo procesar los no etiquetados
            print(f"\n=== Registro {index + 1}/{len(df)} ===")
            print(f"Título: {row['title']}")
            print(f"Géneros: {row['genres']}")
            print(f"Rating: {row['age_rating']}")
            print(f"Descripción: {row['description']}")

            print("\nSelecciona una o más emociones separadas por comas (ej: 9,13):")
            for i, emotion in enumerate(emotion_options, 1):
                print(f"{i}. {emotion}")

            emotion_indices = input("\nEscribe los números de las emociones (o '0' para guardar y salir): ").strip()
            if emotion_indices == '0':
                print("\nGuardando progreso...")
                df.to_csv(file_path, index=False)
                print("Progreso guardado. ¡Hasta luego!")
                break

            try:
                selected_emotions = [emotion_options[int(i) - 1] for i in emotion_indices.split(",") if int(i) > 0 and int(i) <= len(emotion_options)]
                df.at[index, 'emotion'] = ", ".join(selected_emotions)
                print(f"Etiqueta '{', '.join(selected_emotions)}' guardada para el registro {index + 1}.\n")
            except (ValueError, IndexError):
                print("Entrada no válida. Intenta nuevamente con números separados por comas.")

    print("\n¡Todos los registros han sido etiquetados!")
    df.to_csv(file_path, index=False)
    return df

# Llamar a la función para iniciar el etiquetado
dataset = etiquetar_manual(dataset)

```

```python
import pandas as pd

# Ruta del archivo original con etiquetas
file_path = "/content/drive/MyDrive/TFM/Datasets/netflix_sample_for_manual_labeling.xlsx"

# Cargar los datos
data = pd.read_excel(file_path, sheet_name="Movies Dataset")

# Mapeo de números a nombres de emociones
emotion_mapping = {
    1: "Happy", 2: "Sad", 3: "Nostalgic", 4: "Excited", 5: "Scared",
    6: "Angry", 7: "Romantic", 8: "Hopeful", 9: "Funny",
    10: "Adventurous", 11: "Suspenseful", 12: "Inspiring", 13: "Heartwarming"
}

# Función para convertir números a nombres y preservar el orden original
def map_emotions_preserve_order(emotion):
    if isinstance(emotion, str):
        # Dividir y mapear números a nombres respetando el orden
        parts = [part.strip() for part in emotion.split(',')]
        return ", ".join([emotion_mapping[int(part)] if part.isdigit() else part for part in parts])
    return emotion

# Aplicar la transformación
data['emotion'] = data['emotion'].apply(map_emotions_preserve_order)

# Guardar el archivo procesado
output_path = "/content/drive/MyDrive/TFM/Datasets/netflix_sample_ready_for_bert.csv"
data.to_csv(output_path, index=False)

print(f"Archivo con orden preservado guardado en: {output_path}")
```
